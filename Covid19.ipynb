```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("covid19_Confirmed_dataset.csv")
```

```python
df.shape
```

```python
df.dropna(inplace=True,axis=1,how='all')
df.head()
```

```python
df.loc[:, (df != 0).any(axis=0)]
df.head()
```

```python
df.drop(columns=['Lat','Long'],inplace=True)
df.head()
```

```python
df1 = df.groupby(['Country/Region']).sum()
df3 = df1
df1 = df1.reset_index()
df1.head()
```

```python
sns.set()
```

```python
df2 = df1.loc[df1['Country/Region']=='China']
df2 = df2.drop(columns=['Country/Region'])
df2 = df2.T
df2 = df2.reset_index()
df2.columns = ["A","B"]
sns.scatterplot(x="A",y="B",data=df2)
df2.plot(x="A",y="B")
```

```python
df3.loc['China'].plot()
df3.loc['India'].plot()
df3.loc['Afghanistan'].plot()
plt.legend()
```

```python
a=[]
for i in range(df3.shape[0]):
    a.append(df3.iloc[i,:].diff().max())
df3["MaxDiff"] = a
df3
```

```python
df4 = pd.DataFrame(df3.iloc[:,-1])
df4
```

```python
df10 = pd.read_csv('worldwide_happiness_report.csv')
```

```python
df10.head()
```

```python
df11 = df10.drop(columns=['Overall rank','Score','Generosity','Perceptions of corruption'])
df11.head()
```

```python
df11.shape
```

```python
df12 = df11.groupby(['Country or region']).sum()
df12.head()
```

```python
df13 = df12.join(df4)
df13.head()
```

```python
sns.heatmap((df13.corr()))
```
